```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import time
enviornment = {
    "Room1": "clean",
    "Room2": "dirty",
    "Room3": "clean",
    "Room4": "clean",
}
room_pos = {
    "Room1": (0, 1),
    "Room2": (1, 1),
    "Room3": (0, 0),
    "Room4": (1, 0),
}

rooms = list(enviornment.keys())
agent_index = 0

def reflex_agent(state):
    if state == "dirty":
        return "clean"
    else:
        return "Move"

def draw_env(env,agent_pos,step):
    fig,ax = plt.subplots()
    ax.set_xlim(0,2)
    ax.set_ylim(0,2)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title(f"Step: {step} -- Agent Position: {agent_pos}")
    for room,pos in room_pos.items():
        x, y = pos
        color = 'red' if env[room] == "Dirty" else 'green'  
        rect = patches.Rectangle((x, y), 1, 1, facecolor=color, edgecolor='black') 
        ax.add_patch(rect)
        ax.text(x + 0.5, y + 0.5, room, ha='center', va='center', color='white', fontsize=10) # "Room1"
    agent_x, agent_y = room_pos[rooms[agent_pos]] # Finds the agent's current position.
    agent_patch = patches.Circle((agent_x + 0.5, agent_y + 0.5), 0.1, color='blue') #blue circle
    ax.add_patch(agent_patch)

    plt.pause(1) 
    plt.close() 

plt.ion()
steps = 8


for step in range(steps):
    current_room = rooms[agent_index] 
    state = enviornment[current_room] 
    action = reflex_agent(state) 

    draw_env(enviornment, agent_index, step + 1) 

    if action == "Clean": # If the agent decides to clean, this line updates the environment, marking the room as "Clean".
        enviornment[current_room] = "Clean"
    else:
        agent_index = (agent_index + 1) % len(rooms) 


plt.ioff() 
print(" Simulation complete!") 
```

```python
import matplotlib.pyplot as plt

# Environment setup
env_ = {
    "Gate": "Closed",
}

pos_ = {
    "Gate": (1, 1),
}

room = list(env_.keys())
agent_pos_ = (1, 1)  # Starting agent position


def door_agent(state):
    """Toggle the door state"""
    return "Opened" if state == "Closed" else "Closed"


def draw_env_(env_, agent_pos_, step_):
    plt.figure(figsize=(4, 4))
    ax = plt.gca()
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)
    ax.set_xticks([])
    ax.set_yticks([])

    # Draw grid lines
    for x in range(3):
        ax.axvline(x, color='black', lw=1)
        ax.axhline(x, color='black', lw=1)

    # Draw door/Gate
    gate_state = env_["Gate"]
    gate_pos = pos_["Gate"]
    gate_color = 'green' if gate_state == "Opened" else 'red'

    # Draw door as square
    gate_rect = plt.Rectangle((gate_pos[0] - 0.5, gate_pos[1] - 0.5), 1, 1,
                              edgecolor='black', facecolor=gate_color, alpha=0.6)
    ax.add_patch(gate_rect)

    # Add text inside the door
    ax.text(gate_pos[0], gate_pos[1], f"Gate\n{gate_state}", ha='center', va='center', fontsize=9, color='white')

    # Draw agent as a blue circle
    agent_circle = plt.Circle(agent_pos_, 0.2, color='blue')
    ax.add_patch(agent_circle)

    ax.set_title(f"Step {step_}: Door State - {gate_state}, Agent at {agent_pos_}")
    plt.pause(1)
    plt.show()


# === Simulation Loop ===
for step in range(5):  # Simulate 5 steps
    print(f"\nStep {step}")
    
    # Toggle the door
    env_["Gate"] = door_agent(env_["Gate"])
    
    # Draw environment
    draw_env_(env_, agent_pos_=pos_["Gate"], step_=step)
```
